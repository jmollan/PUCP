```python
import requests
from requests.exceptions import ConnectionError, HTTPError, Timeout, RequestException

# La URL a la que se envía la solicitud de búsqueda
url = 'https://e-consultaruc.sunat.gob.pe/cl-ti-itmrconsruc/jcrS00Alias'

# Los datos que se envían con la solicitud POST
data = {
    'accion': 'consPorRuc',
    'razSoc': '',
    'nroRuc': '20100094216',
    'nrodoc': '',
    'token': 'EL_TOKEN_OBTENIDO_DE_LA_PÁGINA',
    'contexto': 'ti-it',
    'modo': '1',
    'rbtnTipo': '1',
    'search1': '20100094216',
    'tipdoc': '1',
    'search2': '',
    'search3': ''
}

# Mejores prácticas: Encabezados que simulan un navegador real
headers = {
    'Content-Type': 'application/x-www-form-urlencoded',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3',
    # 'Cookie': 'TU_COOKIE',
    # 'Referer': 'La URL donde normalmente harías la búsqueda',
    # Más encabezados si son necesarios...
}

# Intentos de conexión
max_retries = 3
for attempt in range(max_retries):
    try:
        # Envía la solicitud POST
        response = requests.post(url, data=data, headers=headers, timeout=10)

        # Asegúrate de que la solicitud fue exitosa
        response.raise_for_status()

        # Si la respuesta es JSON y necesitas extraer información, realiza lo siguiente:
        if response.headers.get('Content-Type') == 'application/json':
            print(response.json()['razonSocial'])  # Asumiendo que 'razonSocial' es una clave válida
            print(response.json()['ruc'])         # Asumiendo que 'ruc' es una clave válida
        else:
            print(response.text)  # O utiliza BeautifulSoup si la respuesta es HTML

        break  # Si todo va bien, rompe el ciclo

    except (ConnectionError, HTTPError, Timeout) as e:
        print(f'Error al enviar la solicitud: {e}. Intento {attempt+1} de {max_retries}.')
        if attempt + 1 == max_retries:
            raise  # Si se alcanza el máximo de intentos, vuelve a lanzar la excepción

    except RequestException as e:
        print(f'Error al enviar la solicitud: {e}.')
        break  # Para otros tipos de excepciones, podrías querer romper el ciclo directamente
```
