```python
from docx import Document

# Carga el documento de Word existente
doc = Document("prueba.docx")

# Crea un nuevo documento de Word
new_doc = Document()

# Copia el contenido al nuevo documento
for element in doc.element.body:
    new_doc.element.body.append(element)

# Guarda el nuevo documento
new_doc.save("no_metadata.docx")
```
